# 블록화를 활용하는 경우

1. 실험조건 외에 실험에 영향을 줄 수 있는 잡음요인을 통제하고자 할 때

2. 동일 조건에서 $2^k$ 요인실험의 모든 시행을 수행하는 것이 불가능한 경우, 비용과 시간을 절약하면서 주요 요인에 대한 신뢰도 높은 결과를 얻고자 할 때.

# 중첩

완전한 요인실험을 블록에 배치하기 위한 설계 기법.  
특정 조합이 특정 블록 내에서만 나타나고 다른 블록에는 나타나지 않는다는 의미이다.  
특정 처리효과를 블록과 구별되지 못하도록 블록효과와 중첩한다.  
요인 간의 변동을 분리해 주요 요인의 효과를 더 정확하게 분석할 수 있다.

## 블록이 2개인 $2^2$ 설계

$AB$가 블록과 중첩:
- $AB$가 $+$ 부호인 모든 처리조합은 <font color='red'>블록1</font>
- $AB$가 $-$ 부호인 모든 처리조합은 <font color='blue'>블록2</font>

애초에 $AB$의 부호를 기준으로 블록을 나누어서 처리효과와 블록은 같은 의미를 가진다. 그래서 $AB$가 블록과 "중첩"되었다고 표현한다.
$$
\text{sign}(AB) \approx \text{block}(data)
$$

블록1 안에서는 $AB$가 $-$일 때의 효과를 알 수 없고, 블록2 안에선 $AB$가 $+$일 때의 효과를 알 수 없다.

## 블록이 2개인 $2^3$ 설계의 4번 반복

$ABC$가 블록과 중첩($ABC$를 기준으로 블록을 구분):
- $ABC$가 $-$ 부호인 모든 처리조합은 블록1
- $ABC$가 $+$ 부호인 모든 처리조합은 블록2

## 블록이 4개인 $2^5$ 설계

**ADE, BCE가 블록과 중첩**
- ADE가 - 부호, BCE가 - 인 모든 처리조합 ➔ <span style="color:red;">블록1</span>
- ADE가 + 부호, BCE가 - 인 모든 처리조합 ➔ <span style="color:blue;">블록2</span>
- ADE가 - 부호, BCE가 + 인 모든 처리조합 ➔ <span style="color:green;">블록3</span>
- ADE가 + 부호, BCE가 + 인 모든 처리조합 ➔ <span style="color:orange;">블록4</span>


두 가지 처리조합의 부호에 따른 파생 조합으로 블록을 나누고 있다.

ADE와 BCE를 중첩시켜서 블록을 나누면 부호상으로는 ABCD를 기준으로 블록을 나눈 것과도 같아진다.

특정 처리조합을 블록에 대응시켜 주요 요인과 효과적으로 분리시켜 더 정확한 분석이 가능하다.

특정 요인을 특정 블록에 대응시키되 다른 요인들의 효과도 충분히, 골고루 반영될 수 있게 하는 방법이다.

## 부분중첩된 $2^3$ 설계

실험 조건이나 자원의 제약으로 인해 일부 요인 조합만 특정 블록 내에서 실험이 가능할 때 활용한다.

반복수1
- ABC가 -인 모든 처리조합 ➔ **블록1**
- ABC가 +인 모든 처리조합 ➔ **블록2**

반복수2
- AB가 +인 모든 처리조합 ➔ **블록1**
- AB가 -인 모든 처리조합 ➔ **블록2**

반복수3
- BC가 +인 모든 처리조합 ➔ **블록1**
- BC가 -인 모든 처리조합 ➔ **블록2**

반복수4
- AC가 +인 모든 처리조합 ➔ **블록1**
- AC가 -인 모든 처리조합 ➔ **블록2**
