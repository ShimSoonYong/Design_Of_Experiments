__반도체 공정__

1. 웨이퍼 제조: 모래에서 규소를 추출해 봉으로 만들고, 그걸 얇게 슬라이스하면 웨이퍼가 된다.
2. 산화 공정: 산화막을 웨이퍼에 씌움
3. 포토공정: 반도체 회로 패턴을 산화막에 새기는 것
4. 식각공정: 반도체 회로 패턴에 맞게 액체나 기체로 웨이퍼와 산화막을 깎아낸다.
5. 증착&이온주입 공정: 증착은 쌓아올리기 위해, 이온을 주입하는 것은 순수한 웨이퍼가 전도성이 없기 때문에 불순물을 섞는 과정이다.
6. 금속배선 공정: 얇은 금속 막을 증착해서 전기가 흐를 수 있게 만드는 것이다.
7. EDS 공정: 양품과 불량품을 구분하는 단계. 수율이 나오는 단계이다.
8. 패키징 공정: 웨이퍼를 잘라서 반도체 칩으로 만들고 기판에 올릴 수 있게 안전하게 포장도 하는 등의 과정을 거쳐 실제 사용이 가능한 반도체 장치가 된다.

과정들을 필요한 만큼 반복한다. 수십일 동안 반복하는 공정도 있다.

웨이퍼가 크면 하나당 생산가능한 반도체 수가 많아 좋고, 얇으면 재료가 덜 들어가 좋다. 그래서 요즘에 쓰는 웨이퍼는 예전 것보다 2배 정도 더 크다.

## 반도체 시장과 고객

반도체의 시작은 트랜지스터, 이후 시장은 계속 성장 중.  
반도체는 설비 하나하나의 가격이 매우 비싸서 아무나 할 수가 없는데다가 사용해야 할 설비도 다양하다. 그 설비들을 잠시라도 가동 정지시키면 회사에겐 유의미한 손해라서 설비가 쉬지 않고 돌아가면서 고장이 나지 않게 공정을 짜야 한다. 그러면서도 주기적으로 정비도 받아 망가지기 전에 고칠 수 있게 스케줄을 짜야 한다.

## 반도체 생산 설비

반도체 생산 설비는 __효율적 운용__ 과 __품질 항상성 관리__ 가 요구되는 핵심 자산이다.  
설비도 비싸지만 기본 원자재인 웨이퍼도 매우 비싸다. 반도체 공장 건물을 건설하는 것보다 그 안에 들어갈 생산 설비가 훨씬 비싸다.

따라서 설비 관리가 핵심 과제 중 하나이다.

자재가 담긴 박스가 레일을 따라 움직이는 층, 실제 생산이 이뤄지는 생산 설비 층, 그리고 그 밑에 안전 설비 층이 있다. 현재는 완전 자동화되어 있다.

## 핵심 기술: 가상화&지능화

반도체 라인이 다 자동화 및 전산화되어 있어서 어떻게 돌아가는지 데이터베이스에서 쉽게 확인할 수 있다.  
지능화는 사람의 지능적 행위를 자동화하는 것. 데이터는 이미 있는 상황이고 유의미한 정보의 추출과 지능화가 과제이다.

설비의 기대 품질과 실제 품질(성능)의 추이는 당연히 차이가 있다. 품질관리는 단일 웨이퍼, 프로세스 대상으로도 해야 하고 설비 하나의 전체 수명주기에 대해서도 해야 한다.

TTTM, Tool to Tool Management. 설비와 설비를 맞추는 품질관리, IBM에서 주창.

각 설비에는 엄청 많은 센서가 있어서 모든 것이 아주 짧은 시간 단위로 데이터로 축적된다.  
너무 많은 데이터를 보관하는 것은 오히려 비용만 키울 수가 있어서 통계 자체를 데이터로 저장하는 것도 방법이다.  

- Trace 데이터: 센서로 측정된 모든 데이터
- Event 데이터: 진행 시간 계산
- Summary 데이터: 통계치
- Alarm 데이터: 문제가 발생했을 때만 저장

웨이퍼 자체도 정밀 가공 제품이기 때문에 계측기로 계측해서 설비를 웨이퍼에 최적화하는 것도 중요하다. 이를 위해 주 공정 전후에 계측을 해서 최적의 공정 파라미터를 통계나 머신러닝 등의 기법으로 원하는 결과에 가깝게 나오도록 미리 세팅하는 시스템을 개발한다.

이러한 자동화된 공정을 돌리다 보면 웨이퍼 별로 지나간 설비와 순서가 달라지게 되고, 이에 따라 최종 칩의 품질도 달라진다. 이를 데이터로 활용해 최적의 품질을 낼 것으로 예측되는 공정 순서를 추천할 수 있다.

같은 종류의 설비끼리도 다소의 성능 차이가 발생한다. 그래서 가장 성능이 좋은 기계의 좋은 결과가 공정 전체로 연쇄되도록 TTTM을 한다.

## 요약

1. 데이터 기반 품질관리는 반도체에서 널리 활용되고 있다.
2. 정보의 전산화가 되었다면, 지능적 행위의 자동화가 가능하다.
3. 정보의 전산화 시, 프로세스 간의 연계성을 고려해야 한다.

반도체에서는 데이터 분석에 들어가는 비용보다 공정 효율 변동에 의한 손해와 이익이 훨씬 크기 때문에 데이터 분석 비용은 별로 중요시되지 않는다.

## 반도체 데이터의 증가

설비에서 데이터가 너무 많이 측정되어 나오는 것도 문제이다. 필요한 정보가 들어있는 데이터를 찾기도 어렵다. 그래서 필요한 데이터를 쉽게 꺼내 써서 볼 수 있는 데이터 구조를 생각하는 것도 중요하고, 데이터 자체의 관리를 잘하는 것도 경쟁력이다.

데이터가 고차원일수록 내포된 정보의 가치가 더 큰 것은 맞다. 그렇다고 무조건 고차원인 것이 좋은 것은 아니다. 문제를 풀기에 적합한 데이터가 무엇인지 알아야 하고, 문제를 풀기에 적합한 데이터는 가능한 한 최대 차원의 데이터가 아닐 수도 있다. 문제에 적합한 데이터의 역할이 분석 기법의 역할보다 훨씬 크다.

강사의 하고 싶은 말:

>분석 기법을 잘 써먹는 것보다 데이터 자체의 특성을 이해하여 적절한 데이터를 추출하고 적합하게 변환하는 것이 중요하다.

교수 질문: 통계 기법과 머신러닝 기법들이 활용되는 분야나 빈도, 어디서 어떻게 쓰이고 있나?  
답변:
- 데이터의 숫자만 가지고서는 데이터를 이해할 수 없다.
- 통계적 기법들을 사용해서 데이터의 특성을 이해해야 한다. 
- 그러고 나서 데이터의 특성과 문제의 성격에 맞게 알고리즘(모델)을 선택해서 사용한다.

학생 질문: 수율 관련 직무를 담당하고 싶은데 학사로는 안되나? 석사를 해야 하나?  
답변: 학석사 수준에선 특정 팀의 특정 직무로 마음대로 들어갈 수 없다. 그것까지 하려면 박사까지 해서 직접 팀에 컨택해야 한다.

질문: 오염되거나 결측치가 많은 데이터가 있을 경우 처리를 어떻게 하고 어떤 방법을 얼마나 신뢰할 수 있나?  
답변:
- 강사는 이미 다 개발된 제품을 잘 돌리는 일을 하기 때문에 보간 같은 걸 해본 적은 없다.
- 하지만 박사논문 때 희소 데이터로 모델을 잘 학습시킬 방법을 고민하기도 했었다.
- 보간법 같은 걸 얼마나 신뢰해야 할지는 정답이 없는 것 같다.

질문: 사용 언어?  
답변:
- 혁신센터에선 SQL, Python, Java, C# 등 7가지 정도를 사용한다.

교수 질문: R, SAS 등의 것들에 대한 회사의 시각  
답변:
- SAS는 비싸서 그냥 Python으로 개발하고 만다.
- R로 전처리나 통계처리를 하는 경우는 꽤 많다. Python과 왔다갔다 한다.

학생 질문: 외부 데이터 결합 시 어떻게 하나?  
답변:
- 강사는 설비 담당이다. 내 입장에서 외부는 공정 데이터일 수 있다. 결국은 회사에서 일하면서 경험을 쌓아야 한다. 데이터만 보고 어떻게 결합할지 알 수 있으면 뛰어난 사람이다.
- 데이터만 보면 되니까 현장을 갈 필요가 없다는 말은 틀린 것은 아니지만, 현장을 직접 체험함으로써 더 잘 이해할 수 있다.
