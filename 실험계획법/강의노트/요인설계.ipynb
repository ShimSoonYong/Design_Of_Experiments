# 요인설계(Factorial Design)

각각의 완전한 시행이나 실험의 반복에서 요인수준의 모든 가능한 조합을 조사하는 것을 의미한다.

교호작용 유무:
- 교호작용이 없으면 A와 B가 상관없이 움직여서 효과에 따른 변화가 평행하다.
- 교호작용이 있으면 A와 B가 상호작용하기 때문에 효과에 따른 변화가 더 역동적으로 변한다.
    - 2요인일 경우에 주효과보다 교호작용이 매우 크면 그래프상으로 반응변수의 추이선이 교차한다.

## 회귀모형 및 반응표면

2요인 요인실험의 회귀모형

$$
y=\beta_0 + \beta_1 x_1 +\beta_2x_2 + \beta_{12} x_1 x_2 +\epsilon
$$

__교호작용항을 배제한 경우__

__교호작용항이 포함된 경우__

교호작용을 포함시키면 잘못된 결론 도출을 피할 수 있다.

## 2요인 요인설계

A요인의 수준이 a, B요인의 수준이 b, 관측이 n번 반복

완전 랜덤화 설계 (Completely Randomized Design)

__효과모형__

$$
y_{ijk}= \mu + \tau_i + \beta_j (\tau\beta)_{ij}+\epsilon_{ijk}\begin{cases}
i=1,2,...,a\\
j=1,2,...,b\\
k=1,2,...,n
\end{cases}
$$

$$
\sum_{i=1}^a \tau_i = 0, \sum_{j=1}^b \beta_j = 0, \sum_{j=1}^b\sum_{i=1}^a (\tau\beta)_{ij}=0
$$

$$
H_0 : \tau_1 = \tau_2 = \cdots = \tau_a = 0 \\
H_1: \text{At least one of } \tau_i \neq 0 \text{ is true}
$$

$$
H_0 : \beta_1 = \beta_2 = \cdots = \beta_b = 0 \\
H_1: \text{At least one of } \beta_i \neq 0 \text{ is true}
$$

$$
H_0 : (\tau\beta)_{ij}=0\ \forall(i,j)\\
H_1: \text{At least one of } (\tau\beta)_{i,j} \neq 0 \text{ is true}
$$

### Two-way ANOVA (Factorial ANOVA)

요인이 2개인 이원분산분석, 교호작용항까지 포함이 가능하다.

- 고정효과모형

$$
\overline{y}_{i..} = \frac{1}{bn} \sum_{j=1}^{b} \sum_{k=1}^{n} y_{ijk} \quad i = 1, 2, \ldots, a
$$

$$
\overline{y}_{.j.} = \frac{1}{an} \sum_{i=1}^{a} \sum_{k=1}^{n} y_{ijk} \quad j = 1, 2, \ldots, b
$$

$$
\sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} \left( y_{ijk} - \overline{y} \right)^2 = bn \sum_{i=1}^{a} \left( \overline{y}_{i..} - \overline{y} \right)^2 + an \sum_{j=1}^{b} \left( \overline{y}_{.j.} - \overline{y} \right)^2 + \\ n \sum_{i=1}^{a} \sum_{j=1}^{b} \left( \overline{y}_{ij.} - \overline{y}_{i..} - \overline{y}_{.j.} + \overline{y} \right)^2 + \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} \left( y_{ijk} - \overline{y}_{ij.} \right)^2
$$

$$
SS_T = SS_A + SS_B + SS_{AB} + SS_E
$$

$$
df \text{ breakdown:}\\ \quad abn - 1 = (a - 1) + (b - 1) + (a - 1)(b - 1) + ab(n - 1)
$$


$$
SS_T = SS_A + SS_B + SS_{AB} + SS_E
$$

$$
SS_T = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} y_{ijk}^2 - \frac{y_{...}^2}{abn}
$$

$$
SS_A = \frac{1}{bn} \sum_{i=1}^{a} y_{i..}^2 - \frac{y_{...}^2}{abn}
$$

$$
SS_B = \frac{1}{an} \sum_{j=1}^{b} y_{.j.}^2 - \frac{y_{...}^2}{abn}
$$

$$
SS_{\text{Subtotals}} = \frac{1}{n} \sum_{i=1}^{a} \sum_{j=1}^{b} y_{ij.}^2 - \frac{y_{...}^2}{abn}
$$

$$
SS_{AB} = SS_{\text{Subtotals}} - SS_A - SS_B
$$

$$
SS_E = SS_T - SS_{AB} - SS_A - SS_B
$$


| Source of Variation | Sum of Squares | Degrees of Freedom      | Mean Square                           | $ F_0 $                |
|---------------------|----------------|-------------------------|---------------------------------------|--------------------------|
| $ A $ treatments  | $ SS_A $     | $ a - 1 $            | $ MS_A = \frac{SS_A}{a - 1} $      | $ F_0 = \frac{MS_A}{MS_E} $ |
| $ B $ treatments  | $ SS_B $     | $ b - 1 $            | $ MS_B = \frac{SS_B}{b - 1} $      | $ F_0 = \frac{MS_B}{MS_E} $ |
| Interaction         | $ SS_{AB} $  | $ (a - 1)(b - 1) $   | $ MS_{AB} = \frac{SS_{AB}}{(a - 1)(b - 1)} $ | $ F_0 = \frac{MS_{AB}}{MS_E} $ |
| Error               | $ SS_E $     | $ ab(n - 1) $        | $ MS_E = \frac{SS_E}{ab(n - 1)} $  |                          |
| Total               | $ SS_T $     | $ abn - 1 $          |                                       |                          |

$3^2$ 요인설계:
- 요인이 2개, 각 요인의 수준이 3개라는 의미.

```python
import numpy as np

Y = np.array([
    [
        [130, 155, 74, 180],
        [34, 40, 80, 75],
        [20, 70, 82, 58]
    ],
    [
[150, 188, 159, 126],
[136,122,106,115],
[25,70,58,45]
    ],
    [
[138,110,168,160],
[174,120,150,139],
[96,104,82,60]
    ]

])
Y
```

```python
Y.sum(axis=2), np.sum(Y.sum(axis=2),axis=0), np.sum(Y.sum(axis=2),axis=1)
```

```python
SSA = 1/12 * np.sum(np.sum(Y.sum(axis=2),axis=0)**2) - np.sum(Y)**2/36
SSB = 1/12 * np.sum(np.sum(Y.sum(axis=2),axis=1)**2) - np.sum(Y)**2/36
SSsubtotal = 1/4 * np.sum(Y.sum(axis=2)**2) - np.sum(Y)**2/36
SSAB = SSsubtotal - SSA - SSB
SSE = np.sum((Y - np.mean(Y))**2) - SSA - SSB - SSAB
SSA, SSB, SSAB, SSE
```

```python
from scipy.stats.distributions import f
import pandas as pd

MSA = SSA / 2
MSB = SSB / 2
MSAB = SSAB/4
MSE = SSE / 27

pd.DataFrame({
    '제곱합': [SSB, SSA, SSAB, SSE, np.sum((Y - np.mean(Y))**2)],
    '자유도': [2,2,4,27,35],
    '평균제곱': [MSB, MSA, MSAB, MSE, None],
    'F-통계량': [MSB/MSE, MSA/MSE, MSAB/MSE, None, None],
    '유의확률': [
                round(1 - f(2, 27).cdf(MSA/MSE),4),
                round(1 - f(2, 27).cdf(MSB/MSE),4),
                round(1 - f(4, 27).cdf(MSAB/MSE),4), None, None]
}, index=['재료 유형', '온도', '교호작용', '오차', '합계'], 
dtype=np.float16)
```

## 3요인 요인 설계 (Three-Factor Factorial Design)
**개요**: 세 개의 독립 요인을 다루며, 요인 A, B, C 각각의 주효과 및 교호작용 효과를 분석한다.

**모형 식**:
$$
Y_{ijkl} = \mu + \tau_i + \beta_j + \gamma_k + (\tau\beta)_{ij} + (\tau\gamma)_{ik} + (\beta\gamma)_{jk} + \\
(\tau\beta\gamma)_{ijk} + \epsilon_{ijkl}\begin{cases}
i = 1,2,...a\\
j=1,2,...,b\\
k=1,2,...,c\\
l=1,2,...,n
\end{cases}
$$

| 변동원인 | 제곱합 ($SS$) | 자유도 ($df$) | 평균제곱 ($MS$) | 기대평균제곱 | $F_0$ |
|:--------:|:-------------:|:-------------:|:---------------:|:------------:|:-----:|
|     A    |     $SS_A$    |     $a - 1$   |     $MS_A$      | $\sigma^2 + \frac{bcn}{a - 1} \sum \tau_i^2$ | $F_0 = \frac{MS_A}{MS_E}$ |
|     B    |     $SS_B$    |     $b - 1$   |     $MS_B$      | $\sigma^2 + \frac{acn}{b - 1} \sum \beta_j^2$ | $F_0 = \frac{MS_B}{MS_E}$ |
|     C    |     $SS_C$    |     $c - 1$   |     $MS_C$      | $\sigma^2 + \frac{abn}{c - 1} \sum \gamma_k^2$ | $F_0 = \frac{MS_C}{MS_E}$ |
|    AB    |    $SS_{AB}$  |  $(a - 1)(b - 1)$  | $MS_{AB}$ | $\sigma^2 + \frac{cn}{(a - 1)(b - 1)} \sum (\tau \beta)_{ij}^2$ | $F_0 = \frac{MS_{AB}}{MS_E}$ |
|    AC    |    $SS_{AC}$  |  $(a - 1)(c - 1)$  | $MS_{AC}$ | $\sigma^2 + \frac{bn}{(a - 1)(c - 1)} \sum (\tau \gamma)_{ik}^2$ | $F_0 = \frac{MS_{AC}}{MS_E}$ |
|    BC    |    $SS_{BC}$  |  $(b - 1)(c - 1)$  | $MS_{BC}$ | $\sigma^2 + \frac{an}{(b - 1)(c - 1)} \sum (\beta \gamma)_{jk}^2$ | $F_0 = \frac{MS_{BC}}{MS_E}$ |
|   ABC    |   $SS_{ABC}$  |  $(a - 1)(b - 1)(c - 1)$  | $MS_{ABC}$ | $\sigma^2 + \frac{n}{(a - 1)(b - 1)(c - 1)} \sum (\tau \beta \gamma)_{ijk}^2$ | $F_0 = \frac{MS_{ABC}}{MS_E}$ |
|   오차   |     $SS_E$    |  $abc(n - 1)$  | $MS_E$ | $\sigma^2$ | - |
|   합계   |    $SS_T$     |  $abcn - 1$   | - | - | - |

기대평균제곱: 오차에 의한 변동과 변인에 의한 변동을 이론적으로 분리한 것

$$
SS_T = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{c} \sum_{l=1}^{n} y_{ijkl}^2 - \frac{y_{....}^2}{abcn}
$$

$$
SS_A = \frac{1}{bcn} \sum_{i=1}^{a} y_{i\cdot\cdot\cdot}^2 - \frac{y_{....}^2}{abcn}
$$

$$
SS_B = \frac{1}{acn} \sum_{j=1}^{b} y_{\cdot j\cdot\cdot}^2 - \frac{y_{....}^2}{abcn}
$$

$$
SS_C = \frac{1}{abn} \sum_{k=1}^{c} y_{\cdot\cdot k\cdot}^2 - \frac{y_{....}^2}{abcn}
$$

$$
SS_{AB} = \frac{1}{cn} \sum_{i=1}^{a} \sum_{j=1}^{b} y_{ij\cdot\cdot}^2 - \frac{y_{\cdot\cdot\cdot\cdot}^2}{abcn} - SS_A - SS_B =\\ SS_{\text{Subtotals}(AB)} - SS_A - SS_B
$$

$$
SS_{AC} = \frac{1}{bn} \sum_{i=1}^{a} \sum_{k=1}^{c} y_{i\cdot k\cdot}^2 - \frac{y_{\cdot\cdot\cdot\cdot}^2}{abcn} - SS_A - SS_C =\\ SS_{\text{Subtotals}(AC)} - SS_A - SS_C
$$

$$
SS_{BC} = \frac{1}{an} \sum_{j=1}^{b} \sum_{k=1}^{c} y_{\cdot jk\cdot}^2 - \frac{y_{\cdot\cdot\cdot\cdot}^2}{abcn} - SS_B - SS_C =\\ SS_{\text{Subtotals}(BC)} - SS_B - SS_C
$$

$$
SS_{ABC} =\\ \frac{1}{n} \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{c} y_{ijk\cdot}^2 - \frac{y_{\cdot\cdot\cdot\cdot}^2}{abcn} - \\SS_A - SS_B - SS_C - SS_{AB} - SS_{AC} - SS_{BC}
= \\SS_{\text{Subtotals}(ABC)} - SS_A - SS_B - SS_C - SS_{AB} - SS_{AC} - SS_{BC}
$$

$$
SS_E = SS_T - SS_{\text{Subtotals}(ABC)}
$$


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.DataFrame({
    'ratio': [10]*8 +  [12]*8 + [14]*8,
    'pressure': ['25psd', '25psd', '30psd', '30psd'] * 6,
    'linespeed': [200, 250] * 12,
    'y': [-3, -1, -1, 1,
          -1, 0, 0, 1,
          0, 2, 2, 6,
          1, 1, 3, 5,
          5, 7, 7 ,10,
          4, 6, 9, 11]
})

df.pivot_table('y', index='ratio', columns=['pressure', 'linespeed'],
               aggfunc='sum', margins=True)
```

```python
SST = df['y'].pow(2).sum() - df['y'].sum()**2/df['y'].count()
SSA = np.sum(df.groupby('ratio')['y'].sum().pow(2)/(2*2*2)) - df['y'].sum()**2/df['y'].count()
SSB = np.sum(df.groupby('pressure')['y'].sum().pow(2)/(3*2*2)) - df['y'].sum()**2/df['y'].count()
SSC = np.sum(df.groupby('linespeed')['y'].sum().pow(2)/(3*2*2)) - df['y'].sum()**2/df['y'].count()
SST, SSA, SSB, SSC
```

```python
SSAB = np.sum(df.groupby(by=['ratio', 'pressure'])['y'].sum().pow(2)/(2*2)) - df['y'].sum()**2/df['y'].count() - SSA - SSB
SSAC = np.sum(df.groupby(by=['ratio', 'linespeed'])['y'].sum().pow(2)/(2*2)) - df['y'].sum()**2/df['y'].count() - SSA - SSC
SSBC = np.sum(df.groupby(by=['pressure', 'linespeed'])['y'].sum().pow(2)/(3*2)) - df['y'].sum()**2/df['y'].count() - SSB - SSC
SSAB, SSAC, SSBC
```

```python
SSABC = np.sum(df.groupby(by=['ratio', 'pressure','linespeed'])['y'].sum().pow(2)/(2)) - df['y'].sum()**2/df['y'].count() - SSA - SSB - SSC - SSAB - SSAC - SSBC
SSABC
```

```python
SSE = SST - SSA - SSB - SSC - SSAB - SSAC - SSBC - SSABC
SSE
```

```python
from scipy.stats.distributions import f

pd.DataFrame({
    '변동원인': ['탄산포화비율(A)', '작용압력(B)', '라인속도(C)',
           'AB', 'AC', 'BC', 'ABC', '오차', '합계'],
    '제곱합': [SSA, SSB, SSC, SSAB, SSAC, SSBC, SSABC, SSE, SST],
    '자유도': [2, 1, 1, 2, 2, 1, 2, 12, 23],
    '평균제곱': [SSA/2, SSB, SSC, SSAB/2, SSAC/2, SSBC, SSABC/2, SSE/12, None],
    'F-통계량': [(SSA/2)/(SSE/12), (SSB/1)/(SSE/12), (SSC/1)/(SSE/12),
             (SSAB/2)/(SSE/12), (SSAC/2)/(SSE/12), (SSBC/1)/(SSE/12),
             (SSABC/2)/(SSE/12), None, None],
    '유의확률': [round(1 - f(2, 12).cdf((SSA/2)/(SSE/12)), 4),
                round(1 - f(1, 12).cdf((SSB/1)/(SSE/12)), 4),
                round(1 - f(1, 12).cdf((SSC/1)/(SSE/12)), 4),
                round(1 - f(2, 12).cdf((SSAB/2)/(SSE/12)), 4),
                round(1 - f(2, 12).cdf((SSAC/2)/(SSE/12)), 4),
                round(1 - f(1, 12).cdf((SSBC/1)/(SSE/12)), 4),
                round(1 - f(2, 12).cdf((SSABC/2)/(SSE/12)), 4), None, None]
})
```

```python
res = sm.OLS.from_formula(formula='y ~ C(ratio) * C(pressure) * C(linespeed)', data=df).fit()
sm.stats.anova_lm(res).round(4)
```

## 2요인 랜덤화 완전블록 요인설계

$$
y_{ijk} = \mu + \tau_i + \beta_j + (\tau\beta)_{ij} + \delta_k + \epsilon_{ijk} \quad
\begin{cases}
i = 1, 2, \ldots, a \\
j = 1, 2, \ldots, b \\
k = 1, 2, \ldots, n \\
\end{cases}
$$

| 변동원인 | 제곱합 | 자유도 | 평균제곱 | $ F_0 $ |
|----------|--------|--------|----------|-----------|
| 블록 | $ \frac{1}{ab} \sum_{k=1}^n y_{..k}^2 - \frac{y_{...}^2}{abn} $ | $ n - 1 $ | $\sigma^2 + ab\sigma^2_{\delta}$ |  |
| A | $ \frac{1}{bn} \sum_{i=1}^a y_{i..}^2 - \frac{y_{...}^2}{abn} $ | $ a - 1 $ | $ MS_A $ | $ \frac{MS_A}{MS_E} $ |
| B | $ \frac{1}{an} \sum_{j=1}^b y_{.j.}^2 - \frac{y_{...}^2}{abn} $ | $ b - 1 $ | $ MS_B $ | $ \frac{MS_B}{MS_E} $ |
| AB | $ \frac{1}{n} \sum_{i=1}^a \sum_{j=1}^b y_{ij.}^2 - SS_A - SS_B - \frac{y_{...}^2}{abn} $ | $ (a-1)(b-1) $ | $ MS_{AB} $ | $ \frac{MS_{AB}}{MS_E} $ |
| 오차 | 차감 | $ (ab-1)(n-1) $ | $ MS_E $ | - |
| 합계 | $ \sum_{i=1}^a \sum_{j=1}^b \sum_{k=1}^n y_{ijk}^2 - \frac{y_{...}^2}{abn} $ | $ abn - 1 $ | - | - |

여기서:
- $ MS_A = \frac{SS_A}{a-1} $
- $ MS_B = \frac{SS_B}{b-1} $
- $ MS_{AB} = \frac{SS_{AB}}{(a-1)(b-1)} $
- $ MS_E = \frac{SS_E}{(ab-1)(n-1)} $


### 예제: 레이더 스코프 목표 탐지 능력 실험

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 데이터 입력
data = {
    'Filter': [1, 2] * 12,
    'Operator': [1, 1, 2, 2, 3, 3, 4, 4] * 3,
    'GroundScatter': ['Low'] * 8 + ['Medium'] * 8 + ['High'] * 8,
    'Intensity': [90, 86, 96, 84, 100, 92, 92, 81,  # 낮은 지면반사산란
                  102, 87, 106, 90, 105, 97, 96, 80,  # 중간 지면반사산란
                  114, 93, 112, 91, 108, 95, 98, 83]  # 높은 지면반사산란
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# ANOVA 모델 생성
model = ols('Intensity ~ C(GroundScatter) + C(Filter) + C(GroundScatter):C(Filter) + C(Operator)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# ANOVA 결과 출력
anova_table
```

## $6\times 6$ 라틴정방설계

$$
y_{ijkl} = \mu + \alpha_i + \tau_j + \beta_k + (\tau \beta)_{jk} + \theta_l + \epsilon_{ijkl}
$$

$$
\begin{cases}
i = 1, 2, \ldots, 6 \\
j = 1, 2, 3 \\
k = 1, 2 \\
l = 1, 2, \ldots, 6 \\
\end{cases}
$$

```python
import pandas as pd

# 데이터 입력 (요인, 오퍼레이터, 필터 수준, 지면 반사 산란 수준, 결과값)
data = {
    '요일': [i for i in range(1, 7, 1) for _ in range(6) ],
    '오퍼레이터': [1, 2, 3, 4, 5, 6] * 6,
    '필터 유형 (f)': 
    ['f1','f1','f1','f2','f2','f2']+
    ['f1','f1','f1','f2','f2','f2']+
    ['f1','f2','f2','f1','f2','f1']+
    ['f2','f2','f1','f1','f1','f2']+
    ['f2','f1','f2','f2','f1','f1']+
    ['f2','f2','f2','f1','f1','f1'],
    '지면 반사 산란 수준 (g)': 
    ['g1', 'g2', 'g3', 'g1', 'g3', 'g2'] + 
    ['g3', 'g1', 'g2', 'g3', 'g2', 'g1']+
    ['g2','g2','g3','g1','g1','g3']+
    ['g2','g1','g1','g2','g3','g3']+
    ['g3','g3','g1','g2','g1','g2']+
    ['g1','g3','g2','g3','g2','g1'],
    '결과값': [
        90, 106, 108, 81, 90, 88,  # 첫 줄
        114, 96, 105, 83, 86, 84,  # 두 번째 줄
        102, 90, 95, 92, 85, 104,  # 세 번째 줄
        87, 84, 100, 96, 110, 91,  # 네 번째 줄
        93, 112, 92, 80, 90, 98,  # 다섯 번째 줄
        86, 91, 97, 98, 100, 92   # 여섯 번째 줄
    ]
}

# 데이터프레임 생성
df = pd.DataFrame(data)
df.tail()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 열 이름의 공백을 언더스코어로 변경
df.rename(columns={'지면 반사 산란 수준 (g)': '지면_반사_산란_수준_g', '필터 유형 (f)': '필터_유형_f'}, inplace=True)

# 수정된 열 이름을 사용하여 모델 정의
model = ols(formula='결과값 ~ C(지면_반사_산란_수준_g) + C(필터_유형_f) + C(지면_반사_산란_수준_g):C(필터_유형_f) + C(요일) + C(오퍼레이터)', data=df).fit()

anova_table = sm.stats.anova_lm(model, type=2)
round(anova_table, 4)
```
